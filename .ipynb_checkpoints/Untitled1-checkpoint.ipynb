```python
df = pd.get_dummies(df, columns=['Brand', 'Model', 'Version', 'Climatisation','Vitres électriques','Sièges électriques', 'Toit'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
y = df['Price']
X = df.drop(['Price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(X , y , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
coef=reg.coef_
coef
```

```python
intercept=reg.intercept_
intercept
```

```python
#Evaluation
from sklearn.metrics import r2_score
y_pred = reg.predict(x_test)
#This metric represents the part of the variance of the dependent variable explained by the independent variables of the model. 
r_squ = r2_score(y_test,y_pred)
r_squ
```

```python
import sklearn
y_pred = reg.predict(x_test)
mse = sklearn.metrics.mean_squared_error(y_test,y_pred)
#MSE It is the average of the squared difference between the predicted and actual value.
mse
```
